```python
from qiskit import *
import numpy as np
```

### Step 1: Building circuits

```python
# Register a quantum register and circuit
q = QuantumRegister(2,'q')
circ = QuantumCircuit(q) 
```

```python
## Add operations to 2 Qubits system

# Add an Hadamard gate H to the 0 qubit and move it to superposition
circ.h(q[0])

# Add an controlled NOT operation between qubit 0 and 1
circ.cx(q[0],q[1])
```

```python
## visualize the circuit
circ.draw()
```

### Step 2: Setup Quantum simulation

```python
from qiskit import BasicAer

## running the above circuit in a statevector simulator
backend = BasicAer.get_backend('statevector_simulator')
```

### Step 3: Running the job

```python
## create a job and execute
job = execute(circ,backend)
```

```python
## printing the job status
job.status()
```

```python
## getting the result from the job
result = job.result()
```

```python
## display the result
output = result.get_statevector(circ, decimals=2)
print(output)
```

```python
## visualize the result 

from qiskit.visualization import plot_state_city
plot_state_city(output)
```
