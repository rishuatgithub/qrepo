```python
import numpy as np

import qiskit
qiskit.__qiskit_version__
```

```python
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
## plot the bloch sphere

plot_bloch_vector([0,0,1], title='Bloch vector')
```

```python
plot_bloch_vector([1,0,0], title='Bloch vector')
```

```python
a = np.array([[1,0,0]])
b = np.array([[0,0,1]])

np.kron(b,a)
```

```python
plot_bloch_vector(np.kron(b,a))
```
